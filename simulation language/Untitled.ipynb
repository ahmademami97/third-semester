```python
# Import Library

import numpy as np 
import matplotlib.pyplot as plt 

# Define Data

team = ['Netprofit','Utilization','makespan','Lab makespan','waiting time']
score=[0.4,0.25,0.15,0.15,0.05]
base = [1, 1.1, 0.9886, 0.3186,0.93]
scenario = [1.1, 1.3054, 0.9930, 0.4955, 0.93]

x_axis = np.arange(len(team))

# Multi bar Chart
plt.figure(figsize=(12,8))
plt.bar(x_axis -0.2, base, width=0.4, label = 'Base model',color='darkcyan')
plt.bar(x_axis +0.2, scenario, width=0.4, label = 'scenario B',color='tomato')
plt.ylabel('\n Normalized Value',fontsize=15)

# Xticks
plt.xticks(x_axis, team,fontsize=12)
plt.xlabel('\n evaluation criteria ',fontsize=15)

# Add legend

plt.legend()

# Display

plt.show()
np.matmul(np.array(score),np.array(base).reshape(-1,1))
```

```python
categories = ['Netprofit','Utilization','makespan','Lab makespan','waiting time']
categories = [*categories, categories[0]]

restaurant_1 = [1, 1.1, 0.9886, 0.3186,0.93]
restaurant_2 = [1.1, 1.3054, 0.9930, 0.4955, 0.93]
restaurant_3 = [1.2, 1.345, 0.9886, 0.4255, 0.89]
restaurant_1 = [*restaurant_1, restaurant_1[0]]
restaurant_2 = [*restaurant_2, restaurant_2[0]]
restaurant_3 = [*restaurant_3, restaurant_3[0]]

label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(restaurant_1))

plt.figure(figsize=(8, 8))
plt.subplot(polar=True)
plt.plot(label_loc, restaurant_1, label='Base Model')
plt.plot(label_loc, restaurant_2, label='Scenario A')
plt.plot(label_loc, restaurant_3, label='Scenario B')
plt.title('Scenarios Comparison', size=20, y=1.05)
lines, labels = plt.thetagrids(np.degrees(label_loc), labels=categories)
plt.legend()
plt.show()
```
